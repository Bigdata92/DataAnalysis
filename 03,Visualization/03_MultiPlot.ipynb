```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic') 
```

## Figure object

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

## Axes object, subplot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2, 1, 1) # 2 X 1 matrix, index = 1 (start index : zero X)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2) # 2 X 1 matrix, index = 2
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221) # (2,2,1) = 221 
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2) # 2 X 2 array (subplots, subplot X)

np.random.seed(2021)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)
fig, axes = plt.subplots(2, 2, figsize=(10,6)) # 2 X 2 array (subplots, subplot X)

for i in range(4):
  ax = axes[i//2, i%2]
  ax.plot(np.random.rand(5))  # row index: i//2, col index: i % 2
  ax.set_title(f"axes {i+1}")
  ax.grid()

fig.suptitle('4개의 그림', fontsize=14)
plt.tight_layout()
plt.show()
```
